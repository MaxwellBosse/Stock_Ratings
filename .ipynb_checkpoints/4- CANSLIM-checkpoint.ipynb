```python
import pandas as pd
import numpy as np
```

```python
ticker = 'DOCU'
```

```python
#Read in earnings data for ticker as df
df = pd.read_csv(f'data/{ticker}_earnings.csv', index_col = 'reportedDate')
df.index = pd.to_datetime(df.index)
```

```python
#Read in Additional CANSLIM inputs from Finviz notebook as df1
df1 = pd.read_csv(f'data/{ticker}_CANSLIM_data.csv', index_col = 'date')
df1.index = pd.to_datetime(df1.index)
```

```python
#Merge dataframes on datetime index
merge = pd.merge(df, df1, how ='outer',left_index = True, right_index = True)
```
